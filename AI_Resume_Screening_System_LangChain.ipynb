```python
#Skill Extraction
from langchain_core.prompts import PromptTemplate

extract_prompt = PromptTemplate(
    input_variables=["resume"],
    template="""
Extract the following details from the resume:
- Skills
- Experience
- Tools

Return in structured format.

Resume:
{resume}
"""
)
```

```python
#Matching
match_prompt = PromptTemplate(
    input_variables=["resume", "job"],
    template="""
Compare the resume with the job description.

Identify:
- Matching skills
- Missing skills

Do NOT assume any skills not present.

Resume:
{resume}

Job Description:
{job}
"""
)
```

```python
#Scoring
score_prompt = PromptTemplate(
    input_variables=["match_result"],
    template="""
Based on the matching result, assign a score between 0 to 100.

Guidelines:
- Strong match → 80-100
- Average match → 50-79
- Weak match → 0-49

Matching Result:
{match_result}
"""
)
```

```python
#Explanation
explain_prompt = PromptTemplate(
    input_variables=["score", "resume"],
    template="""
Explain why this candidate received this score.

Be clear and logical.
Do NOT add any assumptions.

Score: {score}

Resume:
{resume}
"""
)
```

```python
# Job Description

job_description = """
We are hiring a Data Scientist with skills in Python, Machine Learning, SQL, Pandas, NumPy, and Data Visualization.
Experience: 1-3 years
"""

# Strong Resume
resume_strong = """
Skills: Python, Machine Learning, Pandas, NumPy, SQL, Data Visualization
Experience: 2 years in Data Science
"""

# Average Resume
resume_avg = """
Skills: Python, Pandas, Excel
Experience: 1 year in Data Analysis
"""

# Weak Resume
resume_weak = """
Skills: HTML, CSS, JavaScript
Experience: Fresher
"""
```

```python
#Simulated LLM
def fake_llm(prompt):
    return f"[LLM Output]: {prompt}"
```

#### Pipeline functions

```python
#Skill Extraction
def extract(resume):
    prompt_text = extract_prompt.format(resume=resume)
    return fake_llm(prompt_text)
```

```python
#Matching
def match(resume, job):
    prompt_text = match_prompt.format(resume=resume, job=job)
    return fake_llm(prompt_text)
```

```python
#scoring
def score(resume):
    if "Machine Learning" in resume and "SQL" in resume:
        return 85   # Strong
    elif "Python" in resume:
        return 60   # Average
    else:
        return 30   # Weak
```

```python
#Explanation
def explain(score_value, resume):
    prompt_text = explain_prompt.format(score=score_value, resume=resume)
    return fake_llm(prompt_text)
```

```python
#Full Pipeline Function
def pipeline(resume):
    extracted = extract(resume)
    matched = match(resume, job_description)
    score_value = score(resume)
    explanation = explain(score_value, resume)

    return {
        "Extracted": extracted,
        "Matched": matched,
        "Score": score_value,
        "Explanation": explanation
    }
```

```python
print("STRONG:", pipeline(resume_strong))
print("\nAVERAGE:", pipeline(resume_avg))
print("\nWEAK:", pipeline(resume_weak))
```

```python
from langchain_core.runnables import RunnableLambda
```

```python
llm = RunnableLambda(lambda x: f"[LLM Output]: {x}")
```

```python
def extract(resume):
    prompt_text = extract_prompt.format(resume=resume)
    return llm.invoke(prompt_text)

def match(resume, job):
    prompt_text = match_prompt.format(resume=resume, job=job)
    return llm.invoke(prompt_text)

def explain(score_value, resume):
    prompt_text = explain_prompt.format(score=score_value, resume=resume)
    return llm.invoke(prompt_text)
```

```python
print(pipeline(resume_strong))
print(pipeline(resume_avg))
print(pipeline(resume_weak))
```
